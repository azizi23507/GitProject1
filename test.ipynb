```python
arr = np.array([1,2,3,4,5], dtype="U")
print(arr)
print(arr.dtype)
```

```python
#copying the array into a new array and change the data type of array
arr = np.array([1.1, 2.1, 3.1])
print(arr.dtype)

new_arr = arr.astype("i")
print(new_arr.dtype)
```

```python
arr = np.array([1, 0, 3])

newarr = arr.astype(bool)

print(newarr)
print(newarr.dtype)
```

```python
#copying array, when changing the original array, the copied one does not change.
arr = np.array([1, 2, 3, 4, 5])
x = arr.copy()
arr[0] = 42

print(arr)
print(x)
```

```python
# view: make tha view, change the original array, affects the viewed array too
arr = np.array([1,34,3,34,45])
x = arr.view()
arr[0] = 23
print(arr)
print(x)
```

```python
# copies owns the data, and views does not own the data, but we can check this.
# Every NumPy array has the attribute base that returns None if the array owns the data.
# Otherwise, the base  attribute refers to the original object.
arr = np.array([1, 2, 3, 4, 5])
x = arr.copy()
y = arr.view()
print(x.base) # the copy returns None
print(y.base) # the view returns the original array.
```
