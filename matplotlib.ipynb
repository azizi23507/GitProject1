### Display Multiple Plots
With the subplot() function you can draw multiple plots in one figure:

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
#the figure has 1 row, 2 columns, and this plot is the first plot
plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("Sales")
#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
#the figure has 1 row, 2 columns, and this plot is the second plot
plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("Income")


plt.show()
```

```python
# Scatter Plot
"""
The observation in the example above is the result of 13 cars passing by.

The X-axis shows how old the car is.

The Y-axis shows the speed of the car when it passes.

Are there any relationships between the observations?

It seems that the newer the car, the faster it drives, but that could be a coincidence, after all we only registered 13 cars.


"""

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

```python
# check the correlation
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

np.corrcoef(x, y)
# negative correlation one increase the other goes down
```

#### ColorMap
The Matplotlib module has a number of available colormaps.
Used in scatter plots to add a third dimension visually

A colormap is like a list of colors, where each color has a value that ranges from 0 to 100.

Here is an example of a colormap:



This colormap is called 'viridis' and as you can see it ranges from 0, which is a purple color, up to 100, which is a yellow color.

```python
"""
How to Use the ColorMap
You can specify the colormap with the keyword argument cmap with the value of the colormap, in this case 'viridis' which is one of the built-in colormaps available in Matplotlib.

In addition you have to create an array with values (from 0 to 100), one value for each point in the scatter plot
"""
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
# c=colors assigns each dot a value from the colors array
# The dot at x=2, y=111 is index 4 → colors[4] = 40 → teal/blue
# The yellow dot — index 12 → colors[12] = 100 → yellow, at x=6, y=86
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap="viridis")
plt.show()
```

```python
#  can include the colormap in the drawing by including the plt.colorbar() statement:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])


# s=sizes sets each dot's size based on the sizes array
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
# c = y nwo follows the y array instead of a separate array of colors we can also do the same for size
plt.scatter(x, y, c=y, cmap="viridis", s=y)
plt.colorbar()
plt.show()

plt.scatter(x, y, c=y, cmap="viridis", s=sizes)
plt.colorbar()
plt.show()
```

```python
# we can adjust the transparency of the dots with the alpha argument.
x = np.random.randint(100, size=100)
y = np.random.randint(100, size=100)
sizes = 5 * np.random.randint(100, size=100)
colors = np.random.randint(100, size=100)
plt.scatter(x, y, s=sizes, alpha=0.5, c= colors, cmap="viridis")
plt.colorbar()
plt.show()
```

### Creating Bar
with pyplot, we can use the `bar()` function to draw bar graphs. used for categorical data and discrete numerical data


```python
# categorical data
x = np.array(["Blood Type A", "Blood Type B", "Blood Type AB", "Blood Type O"])
y = np.array([30, 25, 10, 35])
plt.bar(x,y)
plt.title("Patient Distribution by Blood Type")
plt.xlabel("Blood Type")
plt.ylabel("Number of Patients")
plt.show()

# discrete numerical data
x = np.array([1, 2, 3, 4])  # number of hospital visits
y = np.array([50, 30, 15, 5])  # number of patients
plt.bar(x, y)
plt.title("Hospital Visit Frequency")
plt.xlabel("Number of Visits")
plt.ylabel("Number of Patients")
plt.show()
```

```python
# to display the bars horizontally instead of vertically, we use the barh() function
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

```python
# The bar() and barh() take the keyword argument color to set the color of the bars:
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x, y, color="red")
plt.show()

# horizontal
plt.barh(x, y, color="#4CAF50")
plt.show()
```

```python
# to set the width of the bar we use the width keyword argument in bar() funciton

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.1, color="pink")
plt.show()

# horizontally
plt.barh(x,y,height=0.1)
plt.show()
```

### Histogram
A histogram is a graph showing frequency distributions. It's used for continues numerical data.

It is a graph showing the number of observations within each given interval.

Example: Say you ask for the height of 250 people, you might end up with a histogram like this:



You can read from the histogram that there are approximately:

2 people from 140 to 145cm
5 people from 145 to 150cm
15 people from 151 to 156cm
31 people from 157 to 162cm
46 people from 163 to 168cm
53 people from 168 to 173cm
45 people from 173 to 178cm
28 people from 179 to 184cm
21 people from 185 to 190cm
4 people from 190 to 195cm

```python
"""
Create Histogram
In Matplotlib, we use the hist() function to create histograms.

The hist() function will use an array of numbers to create a histogram, the array is sent into the function as an argument.

For simplicity we use NumPy to randomly generate an array with 250 values, where the values will concentrate around 170, and the standard deviation is 10.
"""

x = np.random.normal(170, 10, 250)
print(x)
plt.hist(x, color="pink", bins=15)

plt.show()
```

### Pie Charts
By default the plotting of the first wedge starts from the x-axis and moves counterclockwise:



Note: The size of each wedge is determined by comparing the value with all the other values, by using this formula:

The value divided by the sum of all values: `x/sum(x)`

```python
# we use the pie() function to draw pie chats
y = np.array([35, 25, 25, 15])
plt.pie(y)
plt.show()
# the pie chart draws one piece (called a wedge) for each value in the array (in this case [35, 25, 25, 15]).
```

```python
# Add labels to the pie chart with the labels parameter.
# The labels parameter must be an array with one label for each wedge
y = np.array([35, 25, 25, 15])
labels = np.array(["Apples", "Bananas", "Cherries", "Dates"])
plt.pie(y, labels=labels)
plt.show()
```

```python
"""
As mentioned the default start angle is at the x-axis, but you can change the start angle by specifying a startangle parameter.

The startangle parameter is defined with an angle in degrees, default angle is 0
"""
y = np.array([35, 25, 25, 15])
my_labels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, labels= my_labels, startangle= 90)
plt.show()
```

#### Explode
If we want one of the wedges to stand out, the explode parameter allows you to do that.

The explode parameter, if specified, and not None, must be an array with one value for each wedge

```python
y = np.array([35, 25, 25, 15])
my_labels = ["Apples", "Bananas", "Cherries", "Dates"]

my_explode = [0.2, 0, 0, 0]
plt.pie(y, labels=my_labels, explode=my_explode)
plt.show()

my_explode = [0, 0, 0.2, 0.2]
plt.pie(y, labels=my_labels, explode=my_explode)
plt.show()
```

```python
# for making th background white
plt.style.use('default')
plt.rcParams['text.color'] = 'black'
plt.rcParams['axes.labelcolor'] = 'black'
plt.rcParams['xtick.color'] = 'black'
plt.rcParams['ytick.color'] = 'black'
```

```python
# we can add a shadow to the pie chart by setting the shadows parameter to True:
y = np.array([35, 25, 25, 15])
#the bellow to lines are only to make background white so the shadow be visible
my_labels = ["Apples", "Bananas", "Cherries", "Dates"]
my_explode = [0.2, 0, 0, 0]
plt.pie(y, labels=my_labels, explode=my_explode, shadow=True)

plt.show()


```

```python
"""
we an set the color of each wedge with the colors parameter.
The colors parameter, if specified, must be an array with one value for each wedge
'r' - Red
'g' - Green
'b' - Blue
'c' - Cyan
'm' - Magenta
'y' - Yellow
'k' - Black
'w' - White
"""

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show()
```

```python
# we can also add legend
y = np.array([35, 25, 25, 15])
labels = np.array(["Apples", "Bananas", "Cherries", "Dates"])
plt.pie(y, labels=labels, autopct="%1.1f%%")
plt.legend(loc="upper left", title="four fruits:")
plt.show()
```
