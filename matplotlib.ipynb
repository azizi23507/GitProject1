# Matplotlib
matplotlib is a low level graph plotting library in python that serves as a visualization utility.

### plotting

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_points = np.array([0, 5])
y_points = np.array([0, 25])
plt.plot(x_points, y_points, "o")
plt.plot(x_points, y_points)
plt.show()
```

```python
x_points = np.array([1, 2, 6, 8])
y_points = np.array([3, 8, 1, 10])
plt.plot(x_points, y_points, "o")
plt.plot(x_points, y_points)
plt.show()
```

```python
# Default x pints, If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc., depending on the length of the y-points.

y_points = np.array([3, 8, 1, 10, 5, 7,5,7])
plt.plot(y_points)
plt.show()
```

### Markers

```python
# To plot only the markers, you can use shortcut string notation parameter 'o', which means 'rings'.
x_points = np.array([0, 5])
y_points = np.array([0, 25])
plt.plot(x_points, y_points, "o-g")
plt.show()
```

```python
x_points = np.array([1, 2, 6, 8])
y_points = np.array([3, 8, 1, 10])
plt.plot(x_points, y_points, marker="o")
plt.show()
```

### Format Strings `fmt`
`marker|line|color`

```python
y_points = np.array([3, 8, 1, 10])
plt.plot(y_points, "o--y", ms=10) # ms for marker size
```

```python
y_points = np.array([3, 8, 1, 10])

plt.plot(y_points, marker = 'o', ms = 20, mec = 'r', mfc = "y")
# mec color of the edges of the marker amd mfc the inside color
plt.show()
```

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])
y3 = np.array([4,5,3,2,7,8])
plt.plot(y3)
plt.plot(y1)
plt.plot(y2)
plt.xlabel("money spend")
plt.ylabel("time")
```
