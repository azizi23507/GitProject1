```python
from numpy import random
import numpy as np
```

```python
x = random.randint(100) # generate a random integer between 0 and 100
print(x)
print(random.rand()) # random between 0 and 1
print(random.randint(100, size=5)) # generate 1D array with 5 elements between 0 and 100
print(random.randint(100, size=(3, 5))) # 2D array with 3 rows and 5 columns between 0 and 100
print(random.rand(5)) # 1D array with 5 random float elements
print(random.rand(3, 5)) # 2D array, 3 rows, 5 columns, th elements are float.
```

### Generate Random Number From Array
The `choice()` method allows you to generate a random value based on an array of values.<br>
The `choice()` method takes an array as a parameter and randomly returns one of the values.

```python
x = random.choice([3, 5, 7, 9])
print(x)
print(random.choice([1, 3, 5, 7, 9], size=(3, 5)))# to generate a random array of values
```

### Random Distribution

```python
# p is the probability of each of the numbers in the array
x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=100)
print(x)
print(random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(3, 5)))
```

### Shuffling arrays
we can shuffle the elements of the array using the `shufle()` method.

```python
arr = np.array([1, 2, 3, 4, 5])
random.shuffle(arr)
print(arr)
```

### Generating Permutation of Arrays
we can generate the random permutation of the elements of the array using the `permuation()` method.

```python
arr = np.array([1, 2, 3, 4, 5])
print(random.permutation(arr))
```
