### Generate Random Number From Array
The `choice()` method allows you to generate a random value based on an array of values.<br>
The `choice()` method takes an array as a parameter and randomly returns one of the values.

```python
x = random.choice([3, 5, 7, 9])
print(x)
print(random.choice([1, 3, 5, 7, 9], size=(3, 5)))# to generate a random array of values
```

### Random Distribution

```python
# p is the probability of each of the numbers in the array
x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=100)
print(x)
print(random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(3, 5)))
```

### Calculating a Confidence Interval of a Population Proportion
With Python, use the scipy and math libraries to calculate the confidence interval for an estimated proportion.

Here, the sample size is 30 and the occurrences is 6.

```python
import scipy.stats as stats
import math
# Specify sample occurrences (x), sample size (n) and confidence level
x = 6
n = 30
confidence_level = 0.95
# Calculate the point estimate or proportion, the critical z-value, the standard error, and the margin of error
proportion = x / n
critical_z = stats.norm.ppf((1 + confidence_level) / 2)
standard_error = math.sqrt((proportion*(1-proportion)/n))
margin_of_error = critical_z * standard_error

# Calculate the lower and upper bound of the confidence interval
lower_bound = proportion - margin_of_error
upper_bound = proportion + margin_of_error

# Print the results
print("Point Estimate or proportion: {:.3f}".format(proportion))
print("Critical Z-value: {:.3f}".format(critical_z))
print("Margin of Error: {:.3f}".format(margin_of_error))
print("Confidence Interval: [{:.3f},{:.3f}]".format(lower_bound,upper_bound))
print("The {:.1%} confidence interval for the population proportion is:".format(confidence_level))
print("between {:.3f} and {:.3f}".format(lower_bound,upper_bound))
```

### Calculating a Confidence Interval of Student's T Distribution
Here, the sample size is 30, sample mean is 62.1 and sample standard deviation is 13.46.



```python
import scipy.stats as stats
import math

# Specify sample mean (x_bar), sample standard deviation (s), sample size (n) and confidence level
x_bar = 62.1
s = 13.46
n = 30
confidence_level = 0.95

# Calculate, degrees of freedom (df), the critical t-value, and the margin of error
df = n - 1
standard_error = s/math.sqrt(n)
critical_t = stats.t.ppf((1 + confidence_level) / 2, df)
margin_of_error = critical_t * standard_error

# Calculate the lower and upper bound of the confidence interval
lower_bound = x_bar - margin_of_error
upper_bound = x_bar + margin_of_error

# Print the results
print("Critical t-value: {:.3f}".format(critical_t))
print("Margin of Error: {:.3f}".format(margin_of_error))
print("Confidence Interval: [{:.3f},{:.3f}]".format(lower_bound,upper_bound))
print("The {:.1%} confidence interval for the population mean is:".format(confidence_level))
print("between {:.3f} and {:.3f}".format(lower_bound,upper_bound))
```
